# Weighted Term Co-association

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from pathlib import Path
import matplotlib.pyplot as plt
from unsupervised.dual import *
import operator, gensim, joblib
from sklearn.metrics import normalized_mutual_info_score
pd.set_option('display.max_colwidth', 500)
```

### Word Embedding Loading

Load the given word embedding used to construct the similarity matrix S, and calculate coherence of the ensemble topics.

```python
embedding_dir = "/Users/Mark/Desktop/embeddings-w2v/" # path to folder containing embedding
embedding_name = "wikipedia2016-w2v-cbow-d100.bin" # name of word embedding file 
```

```python
embedding_path = Path( embedding_dir ) / embedding_name
embedding = gensim.models.KeyedVectors.load_word2vec_format(embedding_path, binary=True)
print("Loaded embedding - %s" % embedding_path.name)
```

```python
def embedding_similarity( term1, term2, embedding):
    """ Method to calculate similarity between two terms based on a given word embedding"""
    try:
        return embedding.similarity( term1, term2 )
    except:
        return 0
```

### Topic Model Loading

```python
X, all_terms, doc_ids, classes = joblib.load('data/bbc.pkl')
```

```python
dir_results = "data/bbc/nmf_k05/"  # folder containing the topic model rankings and partitions produced by generate-nmf.py
k = 5
kdirname_nmf = "nmf_k%02d" % k

top_terms = 10
num_docs = len(doc_ids)
```

Load the topic models from the specified directory, and truncate the rankings at the specified number of top terms.

```python
def truncate_term_rankings( orig_rankings, top ):
    """ Method to truncate term rankings to the top t terms """
    
    trunc_rankings = []
    if top < 1:
        return orig_rankings
    for ranking in orig_rankings:
        trunc_rankings.append( ranking[0:min(len(ranking),top)] )
    return trunc_rankings
```

```python
rankings = []
partitions = []
for fpath in Path(dir_results).iterdir():
    if fpath.stem.startswith("ranks"):
        (term_rankings,labels) = joblib.load( fpath )
        trunc_rankings = truncate_term_rankings( term_rankings, top_terms)
        rankings.append( trunc_rankings )
    elif fpath.stem.startswith("partition"):
        partition, doc_ids = joblib.load( fpath )
        partitions.append(partition)
print("Loaded %d topic models from %s" % ( len(rankings), dir_results ) )
```

### Weighted Term Coassociation Analysis

Now perform weighted coassociation analysis of the topic models we have just loaded:

```python
wco = WeightedCoassociationModel(k, num_docs, embedding, top_terms )
```

Add all of the models:

```python
for i in range(len(rankings)):
    wco.add( rankings[i], partitions[i])
```

Apply the full matrix building and clustering process:

```python
wco.apply()
```

```python
# displays the final ensemble topics identified by the weighted co-association approach
for term_cluster in wco.term_clusters:
    print(term_cluster)
```

```python
# displays slice of the document-topic assignments generated by the weighted co-association approach
print(wco.partition[:10])
```

### Cluster Evaluation

Create a Data Frame of the results. We include the MDS score of each topic, the least/most coherent pairs in each topic and the individual stability and coherence scores of each topic.

```python
rows = []
for i, term_cluster in enumerate(wco.term_clusters):
    pair_coherences = wco.cluster_pair_coherences[i] # list of dictionaries
    cluster_coherence = wco.evaluate_term_cluster(wco.dfl, term_cluster) # list of coherences
    row = { "Cluster" : (i+1) }
    row["Size"] = len( term_cluster )
    row["Top Terms"] = term_cluster
    row["MDS Score"] = cluster_coherence
    # best and worst pairs?
    ordered_pair_coherences = sorted(pair_coherences.items(), key=operator.itemgetter(1), reverse=True)
    row["Most Coherent Pair"] = sorted(list(ordered_pair_coherences[0][0]))
    row["Least Coherent Pair"] = sorted(list(ordered_pair_coherences[-1][0]))
    row['Stability'] = wco.evaluate_term_cluster(wco.dfc, term_cluster)
    row['Similarity'] = wco.evaluate_term_cluster(wco.dfs, term_cluster)
    rows.append(row)
```

Sort the Data Frame in descending order by cluster MDS score:

```python
df_clusters = pd.DataFrame(rows, columns=["Cluster","MDS Score", "Size","Top Terms","Most Coherent Pair","Least Coherent Pair", "Stability", "Similarity"])
df_clusters = df_clusters.set_index("Cluster")
df_clusters = df_clusters.sort_values( by="MDS Score", ascending=False )
df_clusters
```

### Partition Evaluation

Build a partition of the dataset using the ground truth labels

```python
num_classes = len(classes)
print("Corpus has %d classes" % num_classes )
class_names = list(classes.keys())
class_names.sort()
class_labels = []
doc_map = {}
for i in range( num_docs ):
    class_labels.append( -1 )
    doc_map[ doc_ids[i] ] = i
for class_index in range(num_classes):
    for doc_id in classes[class_names[class_index]]:
        class_labels[doc_map[doc_id]] = class_index
```

Calculate document-partition accuracy using Normalized Mutual Information (NMI)

```python
normalized_mutual_info_score(wco.partition, class_labels, average_method='arithmetic')
```
